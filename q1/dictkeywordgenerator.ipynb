# Create dictionary from mismatched keywords

```python
#process the mismatched keywords
df_mismatch = pd.read_csv('Extra Material 3 - mismatch list.csv')

df_mismatch
```

```python
#if column headers for df_mismatch are found inside the listing name, 
#the column header string will not count as a category match
```

```python
dict_mismatch = dict.fromkeys(df_mismatch, None)
dict_mismatch
```

```python
for keys in dict_mismatch.keys():
    list_miswords = []
    list_miswords.extend(df_mismatch[keys].values)
    dict_mismatch[keys] = list_miswords



def check_mismatch(string):
    string = string.lower().strip()
    for key,value in dict_mismatch.items():
        if len(re.findall('\\b'+key+'\\b', string)) >= 1:
            for var in value:
                if str(var) != None and str(var) in string:
                    string = string.replace(key,'').strip()
                    return string
        else:
            return string

        
dict_mismatch
```

# Process the main item listings

```python
df_keyword_spam_question = pd.read_csv('Keyword_spam_question.csv')

df_keyword_spam_question
```

```python
#remove mismatched keywords from the listings

#match the remaining keywords to the category dictionary and return a list of the matched categories

list_match_categories = []

for index, row in df_keyword_spam_question.iterrows():
    if (index+1) %100000 == 0:
        print(f'processing {index+1}/{len(df_keyword_spam_question)}')
    list_group = []
    listing_name = check_mismatch(row['name'].lower())
    
    for categories in dict_keyword.keys():
        
        if categories in listing_name:

            list_group.append(dict_keyword[categories])
                    
#     print(f'number of categories matched is {len(set(list_group))}')
    list_match_categories.append(list(set(list_group)))

# list_match_categories
```

```python
list_copy = list_match_categories.copy()
#sort the categories returned
for i in range(0,len(list_copy)):
    list_copy[i].sort()

df_keyword_spam_question['groups_found'] = list_copy

df_final = df_keyword_spam_question.drop("name", axis=1)
```
