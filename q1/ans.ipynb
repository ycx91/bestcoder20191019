```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('keywordslist.csv')
```

```python
df
```

```python
df1 = df['Keywords'].str.split(',', expand=True)
```

```python
df1
```

```python
series_temp = df1.unstack()
```

```python
series_temp
```

```python
series_temp = series_temp.str.strip()
series_temp = series_temp.replace('', np.nan)
series_temp = series_temp.dropna()
```

```python
series_temp
```

```python
series_temp.index = series_temp.index.droplevel()
```

```python
series_temp
```

```python
series_temp.name = 'Values'
```

```python
series_temp
```

```python
df
```

```python
df = df.drop('Keywords', axis=1)
```

```python
df = df.join(series_temp)
```

```python
df
```

```python
df = df.reset_index()
```

```python
df = df.drop('index', axis=1)
```

```python
df
```

```python
df = df.drop_duplicates(subset = 'Values')
```

```python
df['len'] = df['Values'].map(str).apply(len)
```

```python
df_ans = df.sort_values(by=['len'], ascending=False)
```

```python
df_ans
```

```python
df_ans.to_csv('keys.csv', index = False)
```
